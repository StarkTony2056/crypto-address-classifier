##Data Collection

In this step, we will basically bring together all the data that we have downloaded from Google Cloud's [BigQuery service](https://console.cloud.google.com/bigquery?p=bigquery-public-data&d=crypto_bitcoin_cash&page=dataset&project=sunlit-center-424211-p6&ws=!1m10!1m4!4m3!1sbigquery-public-data!2scrypto_bitcoin_cash!3sinputs!1m4!4m3!1sbigquery-public-data!2scrypto_dogecoin!3sinputs).

````python
import numpy as np
import pandas as pd

df1 = pd.read_csv('/content/btc-data.csv')
df2 = pd.read_csv('/content/eth-data.csv')
df3 = pd.read_csv('/content/ltc_data.csv')
df4 = pd.read_csv('/content/doge-data.csv')
````

````python
df2.head()
````

````python
btc_df = pd.DataFrame(df1['addresses'])
btc_df.head()
````

````python
doge_df = pd.DataFrame(df4['addresses'])
doge_df.head()
````

````python
lite_df = pd.DataFrame(df3['addresses'])
lite_df.head()
````

````python
df2.columns
````

````python
tmp1 = pd.DataFrame(df2['from_address'])
tmp2 = pd.DataFrame(df2['to_address'])
tmp2.columns = ['from_address']
eth_df = pd.concat([tmp1, tmp2], ignore_index=True)
eth_df.columns = ['addresses']
eth_df.head()
````

````python
ltc_df = lite_df.drop_duplicates(subset='addresses', keep='first')
````

````python
ltc_df.shape
````

````python
ltc_df = ltc_df.head(5000)
````

````python
btc_df['crypto type'] = 'BTC'
eth_df['crypto type'] = 'ETH'
doge_df['crypto type'] = 'DOGE'
ltc_df['crypto type'] = 'LTC'
````

````python
doge_df.head()
````

````python
final_df = pd.concat([btc_df, eth_df, doge_df, ltc_df], axis=0, ignore_index=True)
final_df.head(-5)
````

````python
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)
final_df.head(-5)
````

The final result is saved into a ```.csv``` file.

````python
final_df.to_csv('crypto-dataset.csv', index=False)
````
